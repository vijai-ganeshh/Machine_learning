```python
import seaborn as sns
## Basic ploting with seaborn

tips = sns.load_dataset('tips')
tips
```

```python
#create a scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(x='total_bill',y='tip',data=tips)
plt.title("Scatter plot of total bill vs tip")
plt.show()
```

```python
##line plot
sns.lineplot(x='size',y='total_bill',data=tips)
plt.title("Line plot of Total bill by size")
```

```python
#Categorical plots
#Bar Plot
sns.barplot(x='day',y='total_bill',data = tips)
plt.title('Bar plot of total bill by day')
plt.show()
```

```python
##box plot
sns.boxplot(x='day',y='total_bill',data=tips)
```

```python
##Violin plot(Distribution of above data)
sns.violinplot(x = 'day',y='total_bill',data=tips)
```

```python
##Histograms
sns.histplot(tips['total_bill'],bins=10,kde="True") ##Kernel density estimator = "curve" smoothenes the data 
```

```python
#KDE plot
sns.kdeplot(tips['total_bill'],fill=True)
```

```python
##Pair plot
sns.pairplot(tips) ## Not for categorical only for numerical 
```

```python
## Heat map
corr = tips[['total_bill','tip','size']].corr() ## positive correleation 
sns.heatmap(corr,annot=True,cmap='coolwarm')
corr
```

```python
import pandas as pd
sales_df = pd.read_csv('sales_data.csv')
sales_df.head()
```

```python
##Plot total sales by product
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10,6))
#Put an aggregation function in ""estimator""
sns.barplot(x='Product Category',y='Total Revenue',data=sales_df,estimator=sum)
plt.title('Total Sales of Product')
plt.show()
```
